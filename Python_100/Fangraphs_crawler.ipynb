```python
import os
import requests
from bs4 import BeautifulSoup as BS
```

```python
#all team name to crawl
team_name = 'blue-jays'
url = f'https://www.fangraphs.com/teams/{team_name}'
file = requests.get(url)
soup = BS(file.content,'html5lib')
res = soup.find_all('td',{'class':'frozen'})
res
```

```python
#player Name & playerID & Position & playerURL
import re
team_build = {}

for content in res: 
    P_url = None
    P_pos = None        
    P_id = None
    try:
        P_url = re.search(r'href="(.+)"',str(content)).group(1)
        P_pos = re.search(r'position=(.+)"',str(content)).group(1)        
        P_id = re.search(r'playerid=(.+)&',str(content)).group(1)
    except:
        print('error')
        
    if content.text in team_build.keys():
        continue
    else:
        team_build[content.text] = [P_url,P_pos,P_id]
    
team_build
```
